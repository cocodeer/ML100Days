```python
import pandas as pd
import numpy as np
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
boston = pd.read_csv('boston.csv')
boston.boxplot(figsize=(12,8)) #畫出箱型圖
boston.median().between(300,400) #判斷中位數是否落在300-400
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
boston.plot.scatter(x='NOX', y='DIS',figsize=(12,8))
print("NOX與DIS成反比")
```
